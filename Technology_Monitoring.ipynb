```python
l1=[]
l2=[]
def fun(x):
  if str(x)[0]=='-':
    l1.append(str(x)[1:])
    l2.append('Decrease')
    return (str(x)[1:],"Decrease")
  else:
    l1.append(str(x))
    l2.append('Increase')
    return (str(x),"Increase")
```

```python
for i in range(len(df)):
  fun(round(((df['H'][i]-df['G'][i])/df['G'][i])*100,2))
  
```

```python
df['% age']=l1
df['Change']=l2
```
