```python
import pandas as pd
import numpy as np
```

```python
# Display all columns
df=pd.read_csv("practice.csv")
df.columns.shape[0]
```

```python
# create numerical and categorical columns list
df_numerical=df.select_dtypes(include='number')
df_categorical=df.select_dtypes(exclude="number")
```

```python
# display size of the data-frame
df.shape,df.size
```

```python
# rename column MSSubClass -> SubClass, MSZoning -> Zones
df.rename(columns={"MSSubClass":"SubClass","MSZoning":"Zones"},inplace=True)
```

```python
# display distinct values for Zoning, LotShape, LotConfig
print("Distinct Values of Zoning: ",df["Zones"].unique())
print("Distinct Values of LotShape: ",df["LotShape"].unique())
print("Distinct Values of LotConfig: ",df["LotConfig"].unique())
```

```python
# display count of distinct values for Zoning, LotShape, LotConfig
print(df.groupby(["Zones"])["SubClass"].count(),"\n")
print(df.groupby(["LotShape"])["SubClass"].count(),"\n")
print(df.groupby(["LotConfig"])["SubClass"].count())
```

```python
# max, min of column YearBuilt
print("Max: ",df["YearBuilt"].max())
print("Min: ",df["YearBuilt"].min())
```

```python
# create a new column “year_diff’. This will be holding difference of current year and YearBuilt

df["year_diff"]=df["YearBuilt"].map(lambda x: np.datetime64('today',"Y")-x)
```

```python
# display distinct MSZoning for each OverallQual
df.groupby("OverallQual")["Zones"].unique()
```

```python
# what is maximum LotArea where  BsmtExposure = Mn?
df["LotArea"].where(df["BsmtExposure"]=="Mn").max()
```

```python
# Sort dataframe based on following columns and orders: MSSubClass; ascending, YearBuilt; descending
df.sort_values("SubClass",ascending=True).sort_values("YearBuilt",ascending=False)[["SubClass","YearBuilt"]].head(5)
```

```python
# What is average OverallQual.
df["OverallQual"].mean()
```

```python
# Group by YearBuilt and find maximum OverallQual
df.groupby(["YearBuilt"])[["OverallQual"]].max()
```
